## Первое что мы проверим, на сколько лучше предсказывают модели с заранее претринерованными эмбедингами для юзеров и айтомов с помощью Probabilistic-Matrix-Factorization, проверим это на ddpg

```python
experiments = dict()
for use in [False, True]:
    experiments[use] = experiments.get(use, dict())
    params['pretrain_emb'] = use
    for seed in [1, 2, 3]:
        params['seed'] = seed
        trainer = DRR_Trainer(params)
        experiments[use][seed] = trainer.run_training_loop()
```
